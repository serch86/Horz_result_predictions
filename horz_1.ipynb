```python
# Scrapping
import json
import requests
from bs4 import BeautifulSoup
```

```python
file = "http://www.hipodromo.com.mx/desktop/index.php/carreras/programas"
res = requests.get(file)
html = BeautifulSoup(res.content, 'html.parser')
html
```

```python
#seleccionas seccion completa a scrapear
div = html.find_all("a")
# encuentra los articulos de interes
# ul = div.find_all("div", class_="media border-bottom py-3")
# ul
```

```python
# Recopilar archivos a descargar
links = []
# Se checo a mano de donde iban los programas, pero hay que crear una funcion para descargarlos automaticamente
for ind, line in enumerate(div[30:150]):
    try:
        if line.get('href').split('/')[4] == 'programas':
#             print(ind,line.get('href'))
            links.append('http://www.hipodromo.com.mx'+str(line.get('href')))
    except IndexError:
        continue

links
```

```python
# Descargar pdfs
import requests

for file in links:
    
    name = file.split('/')[-1]
    print(name)
    file_url = file

    r = requests.get(file_url, stream = True) 

    with open(name,"wb") as pdf: 
        for chunk in r.iter_content(chunk_size=1024): 
             # writing one chunk at a time to pdf file 
             if chunk:
                 pdf.write(chunk) 
```
